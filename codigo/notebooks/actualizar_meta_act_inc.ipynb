```python
config_actualizacion_inc = spark.sql("SELECT * FROM datos_origen.config_actualizacion_inc").toPandas()

for _,t in config_actualizacion_inc.iterrows():
    max_id = spark.sql(f'select max({t.columna_max_id}) as max_id from datos_origen.{t.tabla}').first()['max_id']
    if max_id is None:
        max_id = ""
    spark.sql(f"delete from datos_origen.meta_actualizacion_inc where origen = '{t.tabla}'")
    spark.sql(f"insert into datos_origen.meta_actualizacion_inc (origen, max_id) values ('{t.tabla}','{max_id}')")   
```
